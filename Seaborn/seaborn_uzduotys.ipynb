# 1 Atspausdinkite histogramą, kurioje matytųsi, kiek automobilių turi kokią akseleraciją. *Šioje ir kitose užduotyse žaiskite su stiliais ir spalvomis, taip, kaip jums patinka.*

```python
import matplotlib.pyplot as plt
pale = sns.color_palette("flare")
sns.displot(mpg['acceleration'], kde=True)
plt.show()
```

```python
sns.displot(x=mpg['acceleration'], hue=mpg['acceleration'], palette='flare', alpha=1)
```

```python
sns.displot(mpg['acceleration'])
plt.show()
```

# 2 Atspausdinkite histogramą, kurioje matytųsi, kiek automobilių turi kokius variklio tūrius.

```python
sns.displot(mpg['displacement'])
```

# 3 Atspausdinkite histogramą, kurioje matytųsi, kokie yra cilindrų skaičiaus variantai.

```python
sns.countplot(x='cylinders', data=mpg, hue='mpg', palette='flare')
plt.show()
```

```python
sns.countplot(x='cylinders', data=mpg, hue='origin', palette='flare')
plt.show()
```

# 4 Atspausdinkite histogramą, kurioje matytųsi, kiek yra pagaminimo metų variantų

```python
sns.countplot(x='model_year', data=mpg, hue='model_year', palette='flare', legend=False)

plt.show()
```

# 5 Atspausdinkite histogramą, kurioje matytųsi, kiek automobilių lentelėje kokia šalis pagamino.

```python
sns.countplot(x='origin', data=mpg)
```

# 6 Atspausdinkite histogramą, kurioje matytųsi, koks kurioje šalyje pagamintų automobilių variklio tūrio vidurkis.

```python
sns.barplot(x='origin', y='displacement', data=mpg, hue ='origin', palette='flare', legend=False, errorbar=('ci', False))
```

# 6 Atspausdinkite sklaidos diagramą, kurios x ašis būtų 'displacement', y - 'acceleration', taip pat kiekvienas taškas atspindėtų šalį gamintoją ir cilindrų skaičių

```python
sns.scatterplot(x='displacement', y='acceleration', data=mpg, hue='origin', size='cylinders', palette=['#b7ffa8', '#ffc2a8', '#f0a8ff'])
```

# 7 Atspausdinkite visas įmanomas sklaidos diagramas lentelėje, kur pagal taško spalvą matytumėm šalį gamintoją. Kokias tendencijas galima aiškiai išskirti?

```python
sns.pairplot(data=mpg, hue='origin', palette=['#b7ffa8', '#ffc2a8', '#f0a8ff'])
```

# 8 Atspausdinkite stulpelinę diagramą, 'origin' x 'mpg'. Pabandykite interpretuoti rezultatą.

```python
sns.boxplot(x='origin', y='mpg', data=mpg, hue= 'origin', palette=['#b7ffa8', '#ffc2a8', '#f0a8ff'])
```

# 9 Sukurkite koreliacijų matricą. Jos pagrindu atspausdinkite mozaikinę diagramą.

```python
correlation_matrix=mpg.corr(numeric_only=True)
```

```python
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='cividis') #'viridis'
plt.title('Koreliacijų matrica')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Duomenų sąrašai
x = [155, 165, 175, 185, 195, 205]
y = [30, 40, 50, 60, 70, 80]
a = [160, 184, 180, 174]
b = [49, 89, 68, 59]
petras_x = 160  # Petras x koordinatė
petras_y = 80  # Petras y koordinatė
jonas_x = 189
jonas_y = 75

sns.scatterplot(y=y[:len(a)], x=a, label='Ūgiai', color='green')
sns.lineplot(x=x[:len(b)], y=b, label='Svorio linija', color='red')
plt.scatter(petras_x, petras_y, color='#b7ffa8', label='Petras')

sns.scatterplot(x=[jonas_x], y=[jonas_y], label='Jonas', color='black')
#hue='Peras', 'Jonas', palette=['#b7ffa8', '#ffc2a8', '#f0a8ff'])
# Pridėti legendą
plt.legend()

# Pridėti pavadinimą
plt.title('Ūgis ir svoris')

# Pridėti ašių pavadinimus
plt.xlabel('Ūgis')
plt.ylabel('Svoris')

# Rodyti diagramą
plt.show()
```
